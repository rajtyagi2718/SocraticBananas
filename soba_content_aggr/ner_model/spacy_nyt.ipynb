```python
entities = ['quantity', 'unit', 'name', 'comment']
base_data = []  # SENTENCE, {entities : [(START, END, LABEL), (START, END, LABEL)]}
```

```python
# quantity base data

# floats 0.000 ... 1000.000
flts = np.linspace(0, 1000, 1000001).astype(object)
for i in range(0, 1000001, 1000):  # '113.0' -> '113'
    flts[i] = int(flts[i])
for x in flts:
    s = str(x)
    base_data.append([s, {s: [(0, len(s)-1, 'quantity')]}])

# ascii fractions 1/2 1/3 2/3 ... 1/64 ... 63/64
from math import gcd
ascii_fracs = [str(num) + '/' + str(den) 
               for den in range(2, 64) for num in range(den)
               if gcd(num, den) == 1]

# unicode fractions ½ ⅓
unic_fracs = ['⅒', '½', '⅓', '¼', '⅕', '⅙', '⅐', '⅛', '⅑', 
               '⅔', '⅖', '¾', '⅗', '⅜', '⅘', '⅚', '⅝', '⅞']

fracs = ascii_fracs + unic_fracs
for s in fracs:
    base_data.append([s, {s: [(0, len(s)-1, 'quantity')]}])
```

```python
# unit, name

stop_chars = {' ', '/', '\\'}

get_unit_range(ingred, unit):
    """Return start and end indices of unit in ingred."""
    start = ingred.index(unit)
    end = start + len(unit)
    while end < len(ingred) and ingred[end] not in stop_chars:
        end += 1
    end -= 1
    return (start, end)

    
```
